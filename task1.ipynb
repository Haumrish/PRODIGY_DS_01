```python
import matplotlib.pyplot as plt

# Sample data for gender distribution
genders = ['Male', 'Female', 'Non-binary']
counts = [120, 150, 30]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(genders, counts, color=['blue', 'pink', 'purple'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution in a Population')
plt.grid(axis='y')

# Display the chart
plt.show()
```
